## 1) Distribución Weibull

```python
# Distribución Weibull
import numpy as np # importando numpy
from scipy import stats # importando scipy.stats
import pandas as pd # importando pandas
# Muy importante, que los gráficos salgan embebidos en la página.
%matplotlib inline 
# Importar la librería de visualización gráfica
import matplotlib.pyplot as plt # importando matplotlib

tmin, tmax = 1, 1000
t= np.linspace(tmin, tmax, tmax*100)
# media de la muestra seleccionanda al azar
xfSize=13
yfSize=5
plt.figure(figsize=(xfSize, yfSize))

plt.subplot(1, 2, 1);
# parámetros h y c 
h, c = 1, -2
W=(h*(t**(c-1))) * np.exp( - (h/c)*(t**c))
plt.title('Tasa de fallo disminuye \n con el tiempo: h=%f, c=%f'%(h,c))
plt.xlabel('t')
plt.ylabel('f(t)')
_ =plt.plot(t[0:100], W[0:100], linewidth=2, color='r')

plt.subplot(1, 2, 2);
# parámetros h y c 
h, c = 1, 4
W=(h*(t**(c-1))) * np.exp( - (h/c)*(t**c))
plt.title('Tasa de fallo aumenta \n con el tiempo: h=%f, c=%f'%(h,c))
plt.xlabel('t')
plt.ylabel('f(t)')
_ =plt.plot(t[0:1000], W[0:1000], linewidth=2, color='r')
```

## 3) Distribución t de Student

```python
# Distribución t de Student simulada
import numpy as np # importando numpy
from scipy import stats # importando scipy.stats
import pandas as pd # importando pandas
# Muy importante, que los gráficos salgan embebidos en la página.
%matplotlib inline 
# Importar la librería de visualización gráfica
import matplotlib.pyplot as plt # importando matplotlib

np.random.seed(1111111) # para poder replicar los resultados

# media de la muestra seleccionanda al azar
xfSize=12
yfSize=4
plt.figure(figsize=(xfSize, yfSize))

# numero de bins para el histograma
bins=100;

plt.subplot(1, 2, 1);
# parámetros n y muestras
n, muestras = 1, 100000
x_n = np.random.randn(muestras,n) # n columnas de datos normalmente distribuidos (grados de libertad)
x2_n=x_n**2 # calculo todos los cuadrados
Chi_cuadrado=x2_n.sum(axis=1) # suma cada fila para calcular chi-cuadrado y lo pone en una fila, es decir 1D
Chi_cuadrado = np.array([Chi_cuadrado]) # para pasar la fila a un array 2D y calcular la transpuesta
Chi_cuadrado=Chi_cuadrado.T
Chi_cuadrado_n=Chi_cuadrado/float(n)
Square_Chi_cuadrado_n=np.sqrt(Chi_cuadrado_n)
# variable z
z = np.random.randn(muestras,1) # 1 fila de datos normalmente distribuidos
t_student=np.divide(z, Square_Chi_cuadrado_n)

# filtar t_student en valores cercanos a la media, debido a que para n pequeño snunca llega al eje x y 
# el histograma no se pinta
umbral=4.
umbralmin=-umbral
umbralmax=umbral
index= [i for i,v in enumerate(t_student >= umbralmax) if v] # si hay datos mayores de umbralmax saca los 
                                                             # indices para eliminarlos
t_student=np.delete(t_student, index) # array filtrado por el umbral máximo
index= [i for i,v in enumerate(t_student <= umbralmin) if v] # si hay datos menores de umbralmin saca los 
                                                             # indices para eliminarlos
t_student=np.delete(t_student, index) # array filtrado por el umbral mínimo

# pintar el gráfico
# título del gráfico y etiqueta para los ejes x e y
plt.title('Distribucion t-Student:\n n=%g, muestras=%g, media=%g'%(n,muestras, np.mean(Chi_cuadrado)))
plt.xlabel('t-Student')
plt.ylabel('f(t-Student)')
# pintar gráfica t-student
_ =plt.hist(t_student, bins, normed=True, histtype='step',  label='$t_n$')
plt.hold(True)
_ =plt.hist(Square_Chi_cuadrado_n, bins, normed=True, histtype='step', label='$(X_n^2/n)^{0.5}$')
_ =plt.hist(z, bins, normed=True, histtype='step',  label='z')
# leyenda del gráfico
plt.legend()


# segundo gráfico cambiando el parámetro n a 2
plt.subplot(1, 2, 2);
n, muestras = 2, 100000
x_n = np.random.randn(muestras,n) 
x2_n=x_n**2 
Chi_cuadrado=x2_n.sum(axis=1) 
Chi_cuadrado = np.array([Chi_cuadrado]) 
Chi_cuadrado=Chi_cuadrado.T
Chi_cuadrado_n=Chi_cuadrado/float(n)
Square_Chi_cuadrado_n=np.sqrt(Chi_cuadrado_n)
z = np.random.randn(muestras,1) 
t_student=np.divide(z, Square_Chi_cuadrado_n)

umbral=4.
umbralmin=-umbral
umbralmax=umbral
index= [i for i,v in enumerate(t_student >= umbralmax) if v] 
t_student=np.delete(t_student, index)
index= [i for i,v in enumerate(t_student <= umbralmin) if v] 
t_student=np.delete(t_student, index)

plt.title('Distribucion t-Student:\n n=%g, muestras=%g, media=%g'%(n,muestras, np.mean(Chi_cuadrado)))
plt.xlabel('t-Student')
plt.ylabel('f(t-Student)')
_ =plt.hist(t_student, bins, normed=True, histtype='step',  label='$t_n$')
plt.hold(True)
_ =plt.hist(Square_Chi_cuadrado_n, bins, normed=True, histtype='step', label='$(X_n^2/n)^{0.5}$')
_ =plt.hist(z, bins, normed=True, histtype='step',  label='z')
_=plt.legend()
```

## 4) Identificación de un modelo de Poisson de datos muestrales.

```python
# Identificación de una posible Poisson
import numpy as np # importando numpy
from scipy import stats # importando scipy.stats
import pandas as pd # importando pandas
# Muy importante, que los gráficos salgan embebidos en la página.
%matplotlib inline 
# Importar la librería de visualización gráfica
import matplotlib.pyplot as plt # importando matplotlib
from scipy.special import factorial
import scipy.misc #para el factorial vectorizado

muestras=300
precision=0.001
Lambda=1
x_p = np.random.poisson(Lambda,size=(muestras, 1)) # datos distribuidos según una Poisson

# Cálculo de las frecuencias a través de un hitograma de tamaño de bin igual al dato
min=np.amin(x_p)
max=np.amax(x_p)
bins=(max-min)+1

plt.subplot(1, 2, 1);
neventos, posicion_bins, basura = plt.hist(x_p, bins)
plt.title('Poisson $\lambda$=%d'%Lambda)
plt.xlabel('$P(x;\lambda)$')
plt.ylabel('# eventos')

plt.subplot(1, 2, 2);
f_ob=neventos
X= np.linspace(min, max, bins) #espaciado lineal de min a max de uno en uno
Y_x=np.log(f_ob+precision)+np.log(scipy.special.factorial(X))
_=plt.plot(X, Y_x)
```

## Tu turno 2: Demostrar gráficamente como la identificación del modelo de Poisson se realiza de una manera más correcta cuando se cambian los parámetros de la muestra. Ayúdate para demostrar esto del ajuste de una regresión lineal, para obtener la pendiente, ordenada el origen y así la calidad del ajuste.

```python
# obtenemos pendiente, ordenada y la calidad del ajuste para el ejemplo anterior con lambda = 1
import matplotlib.pyplot as plt
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y_x)

print("r-squared:", r_value**2)

plt.plot(X, Y_x, 'o', label='original data')
plt.plot(X, intercept + slope*X, 'r', label='fitted line')
plt.legend()
plt.show()
```

Cuando lambda aumente a valores lo suficientemente grandes, la distribución normal (λ, λ) podría utilizarse para aproximar la distribución de Poisson. La lambda puede ser cualquier valor real positivo. Observemos cómo al ir aumentando el valor de la lambda los valores con más probabilidad se van desplazando hacia la derecha y, al mismo tiempo, los valores posibles se van diversificando.

λ es un parámetro positivo que representa el número de veces que se espera que ocurra el fenómeno durante un intervalo dado. Por ejemplo, si el suceso estudiado tiene lugar en promedio 4 veces por minuto y estamos interesados en la probabilidad de que ocurra k veces dentro de un intervalo de 10 minutos, usaremos un modelo de distribución de Poisson con λ = 10×4 = 40.

Al aumentar lambda se ve como los datos van ajustando en una recta casi perfecta con un gran coeficiente de ajuste.

```python
# Calcular y dibujar la función verosimilitud y soporte para el parámetro μ de la distribución normal de varianza 1.

# parámetros de la normal
mu,s,muestras = 0, 1, 100
# creando la muestra de datos
normal = np.random.normal(mu,s,muestras)

# Valor de los parámetros mínimos y máximos
size = 10
MUmin, MUmax = mu-2, mu+2
MU = np.linspace(MUmin, MUmax, size)
if (s**2)-2 <0:
    S2min, S2max = 0.001, (s**2)+2
else:
    S2min, S2max = (s**2)-2, (s**2)+2
S2= np.linspace(S2min, S2max, size)
MU_bak=MU
S2_bak=S2
# sum_xi-mu2 de la muestra seleccionanda al azar para cada MU
Sxi_mu2=[];
for i in range(MU_bak.shape[0]):
    tmp=(normal-MU_bak[i])**2
    Sxi_mu2.append(tmp.sum())

xfSize=15
yfSize=5
plt.figure(figsize=(xfSize, yfSize))
mediax=MU.mean()

# funcion de verosimilitud  de la muestra extraida
plt.subplot(1, 2, 1);
l_mu_s=(1.0/((np.sqrt(S2)*sqrt(2.0*pi))**muestras))*np.exp((-1.0/(2.0*S2))*(Sxi_mu2))
plt.title('Funcion de verosimilitud Normal n=%d'%muestras)
plt.xlabel('$\mu$')
plt.ylabel('$L(\mu)$')
_ =plt.plot(MU, l_mu_s, linewidth=2, color='r')

# funcion soporte de la muestra extraida
plt.subplot(1, 2, 2);
slx = (1.0/( - muestras*MU) + (np.log(MU) * (muestras*mediax)))
plt.title('Funcion de soporte Normal n=%d'%muestras)
plt.xlabel('x')
plt.ylabel('L(x)')
_ =plt.plot(MU, slx, linewidth=2, color='r')
```
