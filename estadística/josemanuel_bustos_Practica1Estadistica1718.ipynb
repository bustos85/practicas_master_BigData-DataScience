## Tu turno 2: Calcular la media, moda, varianza y coeficiente de variación de las tres distribuciones de propiedades para los cerezos negros.

```python
# obtener los datos del dataframe para cada propiedad sobre los que aplicar los cálculos
from math import pi

datos_altura = dataM[:,2]
datos_diametro = dataM[:,1]/pi
datos_volumen = dataM[:,3]

# media para las tres propiedades. Se obtiene con la función mean()
array_medias = datos_altura.mean(), datos_diametro.mean(), datos_volumen.mean()

# moda para las tres propiedades. Se obtiene con la función de scipy mode()
array_modas = stats.mode(datos_altura,axis=None), stats.mode(datos_diametro,axis=None), stats.mode(datos_volumen,axis=None)

# varianza para las tres propiedades. Se obtiene con la función var()
array_varianzas = datos_altura.var(), datos_diametro.var(), datos_volumen.var()

# coeficiente de variación, sería la división entre la desviación típica y la media
# se puede calcular también con función de scipy variation()
array_coef_scipy = stats.variation(datos_altura), stats.variation(datos_diametro), stats.variation(datos_volumen)
array_coef_manual = datos_altura.std()/datos_altura.mean(), datos_diametro.std()/datos_diametro.mean(), datos_volumen.std()/datos_volumen.mean()

# mostrar todos los arrays con los datos
# medias de altura, diametro y volumen
print("Medias -> Media altura: %0.4f\t Media diametro: %0.4f\t Media volumen: %0.4f\n" % (array_medias))

# modas de altura, diametro y volumen
print("Modas --> Moda altura: %0.4f\t Moda diametro: %0.4f\t Moda volumen: %0.4f\n" % (array_modas[0][0],array_modas[1][0],array_modas[2][0]))

# varianzas de altura, diametro y volumen
print("Varianzas -> Altura: %0.4f\t Diametro: %0.4f\t Volumen: %0.4f\n" % (array_varianzas))

# coeficientes de variación de altura, diametro y volumen
print("Coef.Var.Scipy -> Altura: %0.4f\t Diametro: %0.4f\t Volumen: %0.4f\n" % (array_coef_scipy))
print("Coef.Var.Manual -> Altura: %0.4f\t Diametro: %0.4f\t Volumen: %0.4f\n" % (array_coef_manual))
```

Se puede observar por ejemplo que en las propiedades de altura y diámetro el valor de la media y el valor de la moda son más cercanos, lo que puede indicar que hay mayor concentración de datos en torno a esos valores. En cambio en el volumen la media está alejada del valor de la moda, que es el valor más repetido, esto ocurre porque en la propiedad volumen hay algún dato atípico más alejado en el rango y hace que la media se vea afectada.

Con la varianza se puede ver la dispersión de los datos, y como hemos visto en la propiedad volumen debido a datos atípicos y mayor dispersión de los datos en un rango de valores mayor la varianza es elevada.

## Tu turno 8: Primeramente genera un código que muestre la convergencia de la probabilidad que al lanzar un dado el número que obtiene sea par. Por último calcula y grafica la distribución normal y su función de distribución.

```python
# convergencia de la probabilidad de que al lanzar un dado el número sea par
xfSize=10
yfSize=5
plt.figure(figsize=(xfSize, yfSize))
np.random.seed(2222222)
n=2000
x_u = np.random.rand(1, n) # datos uniformemente distribuidos
# hay dos posibilidades: que el numero obtenido sea par o que sea impar, por lo tanto son dos
# posibles resultados que podemos asignar a 1 o 0
x_u=(x_u>0.5).choose(x_u,1) # cambia los elementos mayor que 0.5 por el evento 1: par
x_u=(x_u<=0.5).choose(x_u,0) # cambia los elementos menores que 0.5 por el evento 0: impar
X=np.zeros((1,n+1),float)
Y=np.zeros((1,n+1),float)
for i in range(1,n+1): Y[0,i]=x_u[0,0:i].sum(); Y[0,i]=Y[0,i]/i; X[0,i]=i;  
plt.title('1. Convergencia de la frecuencia de cara al lanzar un dado n veces')
plt.xlabel('Número de tiradas')
plt.ylabel('Probabilidad')
_ =plt.plot(X[0,1:],Y[0,1:])

# cambiando el valor de la semilla y el número de tiradas
xfSize=10
yfSize=5
plt.figure(figsize=(xfSize, yfSize))
np.random.seed(4444444)
n=4000
x_u = np.random.rand(1, n) # datos uniformemente distribuidos
x_u=(x_u>0.5).choose(x_u,1)
x_u=(x_u<=0.5).choose(x_u,0)
X=np.zeros((1,n+1),float)
Y=np.zeros((1,n+1),float)
for i in range(1,n+1): Y[0,i]=x_u[0,0:i].sum(); Y[0,i]=Y[0,i]/i; X[0,i]=i;  
plt.title('2. Convergencia de la frecuencia de cara al lanzar un dado n veces')
plt.xlabel('Número de tiradas')
plt.ylabel('Probabilidad')
_ =plt.plot(X[0,1:],Y[0,1:])
```

##### comentarios sobre las gráficas de convergencia
Se observa en ambas gráficas como a medida que aumenta el número de tiradas, representado en el eje x de las gráficas, la probabilidad tiende a 0.5 que sería la probabilidad de que al tirar un dado se obtenga un número par, o un número impar, ya que el dado tiene 6 caras y son 3 caras para cada grupo.


```python
# Ahora mostrar la convergencia de la probabilidad utilizando la librería seaborn que muestra 
# varias gráficas según el valor de n, las tiradas realizadas hasta ese momento
import numpy as np
from scipy import stats 
import pandas as pd
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns

xfSize=15
yfSize=15
plt.figure(figsize=(xfSize, yfSize))

dist = stats.beta
n_trials = [5, 10, 20, 40, 75, 100, 150, 250, 500, 1000]

data = stats.bernoulli.rvs(0.5, size=n_trials[-1]) #Generate random numbers, n_trials[-1] es el último de la lista, i.e. 10000
x = np.linspace(0, 1, 100) #espaciado lineal de 100 puntos de 0 a 1

for k, N in enumerate(n_trials):
    sx = plt.subplot(len(n_trials) / 2, 2, k + 1)
    plt.xlabel("$p$, probabilidad de número par")
    plt.setp(sx.get_yticklabels(), visible=False)
    heads = data[:N].sum()
    y = dist.pdf(x, 1 + heads, 1 + N - heads)
    plt.plot(x, y, label="lanzamientos %d,\n %d números pares" % (N, heads))
    plt.fill_between(x, 0, y, color="#348ABD", alpha=0.4)
    plt.vlines(0.5, 0, 4, color="k", linestyles="--", lw=1)
    leg = plt.legend()
    leg.get_frame().set_alpha(0.4)
    plt.autoscale(tight=True)

plt.tight_layout() 
```

##### comentarios sobre las gráficas de convergencia con seaborn
Utilizamos la librería seaborn con la que pintamos varias gráficas, en este caso 10 donde cada gráfica representa un número n de lanzamientos del dado.
Como anteriormente, se puede observar a través de los gráficos como a medida que aumenta el número de tiradas se va hacia la probabilidad de 0,5.


```python
# calcular y graficar la distribución normal
# Graficando Normal
mu, sigma = 0, 0.2 # media y desvio estandar
normal = stats.norm(mu, sigma)
x = np.linspace(normal.ppf(0.01),
                normal.ppf(0.99), 100)
fp = normal.pdf(x) # Función de Probabilidad
plt.plot(x, fp)
plt.title('Distribución Normal')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()

fd = normal.cdf(x) # Función de Probabilidad
plt.plot(x, fd)
plt.title('Distribución Normal')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

##### comentarios sobre la distribución normal y su función de distribución
La distribución normal es una distribución continua que tiene forma acampanada, y es simétrica respecto a su media.
